```python
# Key analysis notebook content
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json

# Load results
with open('../results/raw/latest_results.json', 'r') as f:
    results = json.load(f)

df = pd.json_normalize(results)

# Performance analysis
performance_by_tool = df.groupby('tool').agg({
    'execution_time': ['mean', 'std', 'count'],
    'success': 'mean'
})

print("Performance Summary:")
print(performance_by_tool)

# Visualization
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='tool', y='execution_time')
plt.title('Execution Time Distribution by Tool')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
